## (01/01/2026) Day 1: Temperature Converter
**(C to F and F to C)**<br>
**Logic:** C to F - Formula used $F = (C\times 9/5) + 32$<br>
**Logic:** F to C - Formula used $C = (F - 32)\times 5/9$


```python
def CToF(Celcius):
  '''Converts Celcius to Fahrenheit'''
  Fahrenheit = (Celcius * 9/5) + 32
  return Fahrenheit

def FToC(Fahrenheit):
  '''Converts Fahrenheit to Celcius'''
  Celcius = (Fahrenheit - 32) * 5/9
  return Celcius

# User Interaction
print("Temperature Converter")
print("----------------------")
print('Enter Type of Conversion')
print('1. Celcius to Fahrenheit')
print('2. Fahrenheit to Celcius')

# Input Validation
try:
  choice = int(input('Enter Choice (1 or 2): '))
  if choice == 1 or choice == 2:
    temp = float(input('Enter Temperature: '))
    if choice == 1:
      print(f'{temp} Celcius is {CToF(temp)} Fahrenheit')
    if choice == 2:
      print(f'{temp} Fahrenheit is {FToC(temp)} Celcius')
  else:
    print('Invalid Choice')
except ValueError:
  print('Invalid Input')
```

## (02/01/2026) Day 2: Even, Odd or Prime Number Identifier.
**Objective:** To identify if a number is even/odd and/or Prime<br>
**Logic:**<br>
Even/Odd - whether divisible by 2?<br>
Prime - <br>
a) Not 0 or 1 <br>
b) Not multiple of 2 or 3 <br>
c) No divisor till $sqrt(n)$


```python
import math

def isEven(num):
  '''Identifies a number as Odd or Even'''
  if num % 2 == 0:
    return True
  else:
    return False

def isPrime(num):
  '''Identifies a number as Prime or Not Prime'''
  # Numbers 0 & 1 are Non-prime
  if num <= 1:
    return False
  # Numbers 2 & 3 are Prime
  if num <= 3:
    return True
  # Numbers multiples of 2 and 3 cannot be Prime!
  if num % 2 == 0 or num % 3 == 0:
    return False
  # Look for divisors for a number till sqr root of the number
  limit = int(math.sqrt(num)) + 1
  # All numbers before 5 are accounted for.
  # the number and number+2 are CHECKED
  # A step of 4 is used as 1 & 3 after 5 are even
  for i in range(5, limit, 4):
    if num % i == 0 or num % (i+2) == 0:
      return False
  return True

# User Interaction
myNum = int(input('Enter a number: '))

# Input Validation
try:
  if isEven(myNum):
    print(f'{myNum} is an Even Number')
  else:
    print(f'{myNum} is an Odd Number')
  if isPrime(myNum):
    print(f'{myNum} is a Prime Number')
except ValueError:
  print('Invalid Input')
```

## (03/01/2026) Day 3: FizzBuzz Logic
**Objective:** Handle Multiple Conditions<br>
**Logic:** <br>
**1.** If $n$ is divisible by 3 & 5 $\rightarrow$ "FizzBuzz"<br>
**2.** If $n$ is divisible by 3 $\rightarrow$ "Fizz"<br>
**3.** If $n$ is divisible by 5 $\rightarrow$ "Buzz"<br>
**4.** If $n$ is divisible by 3 & 5 $\rightarrow$ print $n$

```python
def FizzBuzz(num):
  '''Looks for the number to be divisible by 3, 5 or both and
  returns appropriate response'''
  if num % 3 == 0 and num % 5 == 0:
    response = 'FizzBuzz'
  elif num % 3 == 0:
    response = 'Fizz'
  elif num % 5 == 0:
    response = 'Buzz'
  else:
    response = num
  return response

# User interaction
myNum = int(input('Enter number till you want to see FizzBuzz: '))

# Input validation
try:
  for i in range(1, myNum + 1):
    print(FizzBuzz(i))
except ValueError:
  print('Invalid Input')
```

##(04/01/2026) Day 4: Professional Tip Calculator

**Logic:**<br>
1) Use input to get Total bill, Tip% and No. of people <br>

```python
print('___Professional Tip Calculator___')
print()
def calcTip(bill, tipPercent):
  '''Calculates tip based on bill and tip percentage'''
  tip = bill * (tipPercent/100)
  return tip

def calcSplit(bill, pax):
  '''Calculates split based on bill and number of persons'''
  split = bill / pax
  return split

# User interface
myBill = float(input('What is the total bill? $'))
myTip = float(input('What percentage tip do you prefer(10, 12, 15, 20)? '))
myPax = int(input('How many people to split the bill? '))

# Input validation
try:
  tip = calcTip(myBill, myTip)
  # The tip must be included in the amount to be split!!
  totalBill = myBill + tip
  split = calcSplit(totalBill, myPax)
  print(f'As per your instructions, the tip comes to: ${tip:.2f}')
  print(f'The total bill comes to: ${totalBill:.2f}')
  print(f'Each person should pay: ${split:.2f}')
except ValueError:
  print('Invalid Input')
```

**Objective:** Master While loop and Break statement
**Logic:** Maintain a loop as long as the user has attempts left. Use 'break' to exit early upon success or conclude with a lockout message if attempts are exhausted.

## (06/01/2026) Day 6: String Manipulation - Reverse a sentence and Count Vowels
**Objective:** Explore three methods to reverse a string AND count Vowels<br>
**Logic:** <br>
**1.** Reversing a sentence using, slicing method, reversed iterator and for loop. <br>
**2.** Counting number of Vowels and identifying how many times a vowel is repeated in the sentence. <br>

```python
def isVowel(x):
  '''Identifies Vowels'''
  vowels = ['a', 'e', 'i', 'o', 'u']
  if x.lower() in vowels:
    return True
  else:
    return False

myString = 'The brown fox jumps over a lazy dog in the moonlit night'
print(myString)
print()

# Using slice method - standard and fastest
revString_slice = myString[::-1]
print(f'String Reversal using splice: {revString_slice}')

# Using reversed iterator - large dataset - memory efficient (does not duplicate data)
revIterator = reversed(myString)

# Using .join() to get back the string
revString_reversed = ''.join(revIterator)
print(f'String Reversal using reversed iterator: {revString_reversed}')

# Using for loop
revString_for = ''
vowelCount = 0
# Creating a dictionary with vowels as keys and count as values
vowelFrequency = {}
for v in 'aeiou':
  vowelFrequency[v] = 0

for i in range(len(myString), 0, -1):
  revString_for += myString[i-1]
  if isVowel(myString[i-1]):
    vowelCount += 1
    vowelFrequency[myString[i-1]] += 1

print(f'String Reversal using for loop: {revString_for}')
print(f'Number of vowels in the sentence: {vowelCount}')
print(f'Vowel Frequency: {vowelFrequency}')
```

## (08/01/2026) Day 8: Multi-Table Grid Generator (Nested Loops)
Objective: Master nested for loops and advanced string alignment.
Logic: An outer loop iterates through rows, while an inner loop iterates through columns. We calculate the product and use f-string padding {:4d} to ensure a clean, tabular display.

```python
def multiTable():
  ''' Generates Multiplication Table '''
  print('___Multiplication Table Generator___')
  print()
  try:
    x = int(input('Enter number of Rows: '))
    y = int(input('Enter number of Columns: '))
    print(f'\nGenerating a {x} x {y} multiplication table:')
    print()

    # Outer loop for Rows
    for i in range(1, x + 1):
      # Inner loop for Columns
      for j in range(1, y + 1):
        # Product i * j
        product = i * j
        # Print product followed by 4 spaces for alignment
        # Using "end=' '" prevents moving to a new line
        print(f'{product:4d}', end=' ')

      # Separator between Rows
      print()
  except ValueError:
    print('Error! You did not enter a valid number')

multiTable()
```

## (09/01/2026) Day 9: Guess the Number Game
**Objective:** Learn to import modules and manage game-state logic. <br>
**Logic:** Use the random library to generate a target. Implement a while loop to keep the game running untill 10 attempts and use conditional logic to provide "High/Low" feedback to the user.

```python
def checkNum(x, y):
  '''Check if 2 numbers are equal, lower or higher'''
  if x == y:
    print('Congratulations! You guessed correctly')
  elif x > y:
    print('Too High')
  else:
    print('Too Low')

def hintNum(x, y):
  '''Divisibility Hint'''
  if y % 2 == 0:
    print('Hint: The number is an Even!')
  if y % 3 == 0:
    print('Hint: The number is a Multiple of 3')
  if y % 5 == 0:
    print('Hint: The number is a Multiple of 5')
  if y % 7 == 0:
    print('Hint: The number is a Multiple of 7')
  if y % 11 == 0:
    print('Hint: The number is a Multiple of 11')
  else:
    print('Sorry! No Hint Available')


def guessNum():
  ''' Guess the Number Game '''
  print('__Guess the Number Game__')
  print()
  import random
  secretNum = random.randint(1, 101)
  attempts = 0
  while attempts <= 10:
    try:
      guess = int(input('Enter your guess: '))
      attempts += 1
      checkNum(guess, secretNum)
      if guess == secretNum:
        print(f'You guessed it in {attempts} attempts')
        break
      if attempts > 6:
        hintNum(guess, secretNum)
      if attempts == 10:
        print('Sorry! You have exhausted all your attempts')
        print(f'The number I had thought of was {secretNum}')
    except ValueError:
      print('Invalid Input')

guessNum()
```

## (10/01/2026) Day 10: Rock-Paper-Scissors Game
***Objective:*** Handle multi-variable comparisons and list-based randomization.<br>
***Logic:***<br>
 - Utilize random.choice() for computer selection and a consolidated if statement with logical or operators to determine the winner efficiently.<br>
 - while loop for best of five!

```python
import random

def playRPS():
  options = ['rock', 'paper', 'scissors']
  myScore = 0
  computerScore = 0
  print('__Rock Paper Scirrors Showdown__')
  print('Best of 5')
  print()
  # while loop for best of 5
  while myScore < 3 and computerScore < 3:
    # User input - case insensitive
    myChoice = input('Enter your choice (rock/paper/scissors): ').lower().strip()
    if myChoice not in options:
      print('Invalid Input!')
      return

    computerChoice = random.choice(options)
    print(f'Computer chose {computerChoice}')

    if myChoice == computerChoice:
      print('It\'s a tie!')
    elif myChoice == 'rock' and computerChoice == 'scissors':
      print('You win!')
      myScore += 1
    elif myChoice == 'rock' and computerChoice == 'paper':
      print('You lose!')
      computerScore += 1
    elif myChoice == 'paper' and computerChoice == 'rock':
      print('You win!')
      myScore += 1
    elif myChoice == 'paper' and computerChoice == 'scissors':
      print('You lose!')
      computerScore += 1
    elif myChoice == 'scissors' and computerChoice == 'paper':
      print('You win!')
      myScore += 1
    elif myChoice == 'scissors' and computerChoice == 'rock':
      print('You lose!')
      computerScore += 1

    print(f'Your Score: {myScore}')
    print(f'Computer Score: {computerScore}')
    print()

  if myScore > computerScore:
    print('Congratulations! You won the game!')
  else:
    print('Sorry! You lose!')

# Execute Game
playRPS()
```

## (11/01/2026) Day 11: Palindrome Checker (String Slicing)
**Objective:** Master string normalization and the [::-1] slicing technique.<br>
**Logic:** 1. Clean the input by removing non-alphanumeric characters and converting to lowercase.<br>
2. Use Python's slicing notation string[start:stop:step] with a step of -1 to reverse the string.<br>
3. Compare the original cleaned string with the reversed version.

```python
def textCleaner(text):
  '''Removes spaces and converts all to lowercase'''
  return ''.join(char for char in text if char.isalnum()).lower()
# List Comprehension

def reversal(string):
  '''Reverses string'''
  return string[::-1]

# User Interface
print('__Palindrome Checker__')
print()
myText = input('Please Enter Your String: ')

cleaned = textCleaner(myText)
reversed = reversal(textCleaner(myText))
print(f'Original Text without spaces: {cleaned}')
print(f'Reversed Text without spaces: {reversed}')

if cleaned == reversed:
  print('The string is a Palindrome')
else:
  print('This string ain\'t a Palindrome')
```

## (12/01/2026) Day 12: Leap Year Checker
**Goal:** To identify whether the entered year is a Leap Year and if it is NOT, then to find the last and the next leap year! <br>
**Objective:** Master nested if-else structures and hierarchical logic.<br>
**Logic:** A year is a leap year if it is divisible by 4, except for years divisible by 100, unless those years are also divisible by 400. <br>
This requires three levels of nested conditions to evaluate correctly.<br>

```python
def isLeap(year):
  '''Checks if entered year is a Leap Year'''
  if year % 4 == 0:
    if year % 100 != 0:
      return True
    else:
      if year % 400 == 0:
        return True
  return False

# User Interface
print('__Leap Year Checker__')
print()
try:
  myYear = int(input('Enter Year: '))
  if isLeap(myYear):
    print(f'{myYear} is a Leap Year')
  else:
    print(f'{myYear} is not a Leap Year')
    myYearNxt = myYear
    while isLeap(myYearNxt) == False:
      myYearNxt += 1
      if isLeap(myYearNxt):
        print(f'{myYearNxt} is the next Leap Year')
    myYearLast = myYear
    while isLeap(myYearLast) == False:
      myYearLast -= 1
      if isLeap(myYearLast):
        print(f'{myYearLast} was the last Leap Year')
except ValueError:
  print('Invalid Input')
```

## (13/01/2026) Day 13: Password Strength Checker
**Objective:** Master string iteration and character-type analysis.<br>
**Logic:** 1. Initialize flags for length, numbers, uppercase, and special symbols.<br>
2. Iterate through the string to validate each character.<br>
3. Calculate a "Strength Score" based on met criteria and provide feedback.<br>

```python
def checkPasswordStrength(password):
  '''Checks Password Strength'''
  # Flags
  has_Length = len(password) >= 8
  has_Digit = False
  has_Upper = False
  has_Lower = False
  has_Special = False

  specialChar = '~!@#$%&*+'

# Iterate and Check Flags
  for char in password:
    if char.isdigit():
      has_Digit = True
    if char.isupper():
      has_Upper = True
    if char.islower():
      has_Lower = True
    if char in specialChar:
      has_Special = True
  # Evaluating Password and Calculating Strength
  passwordStrength = [has_Length, has_Digit, has_Upper, has_Lower, has_Special]
  passwordScore = sum(passwordStrength)
  print(passwordStrength)

  # Output
  print('__Professional Password Analyzer__')
  print()
  print(f'---Password Analysis for: {'*' * len(password)}')

  if passwordScore == 5:
    print('Password is Strong')
  elif passwordScore == 4 or passwordScore == 3:
    print('Password is Moderate')
  else:
    print('Password is Weak')

# User Interface
myPassword = input('Enter Password: ')
checkPasswordStrength(myPassword)
```

## (13/01/2026) Password Strength Checker - Upgrade
**Upgrade:** Suggests modification to current Password

```python
def checkPasswordStrength(password):
  # Flags (List Comprehension)
  has_Length = len(password) >= 8
  has_Digit = any(char.isdigit() for char in password)
  has_Upper = any(char.isupper() for char in password)
  has_lower = any(char.islower() for char in password)
  has_special = any(char in '~!@#$%&*_+' for char in password)

# List of Tuples (Flag, what is required)
  requirements = [
      (has_Length, 'Must be at least 8 characters'),
      (has_Digit, 'Must contain at least one digit'),
      (has_Upper, 'Must contain at least one uppercase'),
      (has_lower, 'Must contain at least one lowercase'),
      (has_special, f'Must contain at least one special character {'~!@#$%&*_+'}')
  ]
  # Calculating Strength Score - counts True in each tuple
  score = sum(flag for flag, desc in requirements)
  # User Interface
  print('__Professional Password Analyzer__')
  print()
  print(f'---Password Analysis for: {'*' * len(password)}')
  if score == 5:
    print('Password is Strong')
    print('No changes required')
  if score == 4 or score == 3:
    print('Password is Moderate')
  if score <= 2:
    print('Password is Weak')

  for flag, desc in requirements:
    if not flag:
      print(f'Recommended Change: {desc}')

# User Interface
myPassword = input('Enter Password: ')
checkPasswordStrength(myPassword)
```

## (14/01/2026) Day 14: Interactive Shopping List (List Basics)
**Objective:** Master List manipulation including adding, viewing, and removing elements.<br>
**Logic:** Use a while loop to maintain an interactive session. Utilize .append() for additions and enumerate() for indexed viewing. Implement error handling for removal of non-existent items.

```python
def shopping_list():
  shoppingList = []

  while True:
    print('__Smart Shopping Manager__')
    print()
    print('1. View List')
    print('2. Add Item')
    print('3. Remove Item')
    print('4. Exit')
    choices = [1, 2, 3, 4]
    # User Interface
    myChoice = int(input('Choose an option (1-4): '))
    if myChoice not in choices:
      print('Invalid Input')
      continue
    if myChoice == 1:
      if shoppingList == []:
        print('Empty Shopping List!!')
      else:
        print('Your Shopping List: ')
        for index, item in enumerate(shoppingList, start=1):
          print(f'{index}. {item}')
    elif myChoice == 2:
      newItem = input('Enter item: ').strip().capitalize()
      shoppingList.append(newItem)
      print(f'{newItem} successfully added to the list')
    elif myChoice == 3:
      if shoppingList == []:
        print('Empty Shopping List!!')
      else:
        print('Current Shopping List for your reference:')
        for index, item in enumerate(shoppingList, start=1):
          print(f'{index}. {item}')
        removeItem = int(input('Enter the index of the item to remove: '))
        if removeItem < 1 or removeItem > len(shoppingList):
          print('Invalid Index')
        else:
          removedItem = shoppingList.pop(removeItem - 1)
          print(f'{removedItem} successfully removed from the list')
    elif myChoice == 4:
      for index, item in enumerate(shoppingList, start=1):
        print(f'{index}. {item}')
      print('Exiting Program')
      break

shopping_list()

```

## (15/01/2026) Day 15: Grand Review - Advanced Multi-Input Calculator
**Objective:** Master variable-length arguments (*args) and list aggregation.<br>
**Logic:** Use *args to create functions that are not limited to two inputs. Implement a collection loop that builds a list of numbers dynamically before passing them to the math functions using "argument unpacking."

```python
def add(*args):
  return sum(args)

def multiply(*args):
  result = 1
  for n in args:
    result *= n
  return result

def average(*args):
  return sum(args) / len(args)

def advanced_calc():
  print('__Phase 1 Capstone: Multi Input Calculator__')
  print()
  numbers = []
  # User Interface (1): Generate list of numbers (*args)
  while True:
    val = input('Enter a number (or type"done" to calculate): ').lower().strip()
    if val == 'done':
      break
    try:
      numbers.append(float(val))
    except ValueError:
      print('Invalid Input')

  # User Interface (2): Operation Input
  print('Select an Operation')
  print('1. Addition')
  print('2. Multiplication')
  print('3. Average')
  myChoice = int(input('Enter your choice (1/2/3): '))
  Choices = [1, 2, 3]
  if not numbers:
    print('No numbers entered!! Exiting')
    return
  if myChoice not in Choices:
    print('Invalid Input')
    return

  # Syntax (*numbers) unpacks list numbers into arguments for the function
  operation = ''
  if myChoice == 1:
    operation = 'Addition'
    result = add(*numbers)
  elif myChoice == 2:
    operation = 'Multiplication'
    result = multiply(*numbers)
  elif myChoice == 3:
    operation = 'Average'
    result = average(*numbers)
  print(f'Result of {operation}: {result:.2f}')

advanced_calc()
```
